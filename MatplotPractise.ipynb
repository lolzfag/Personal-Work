```python
import os
import numpy as np
import matplotlib.pyplot as plt

x = [u'Diabetes', u'Heart Disease',u'Total Market' ]
y = [536,729,9930]

fig, ax = plt.subplots()    
width = 0.75 # the width of the bars 
ind = np.arange(len(y))  # the x locations for the groups
ax.barh(ind, y, width, color="blue")
ax.set_yticks(ind+width/2)
ax.set_yticklabels(x, minor=False)
plt.title('TAM in Europe')
plt.xlabel('TAM in millions')
plt.ylabel('Type of Markets') 
for i, v in enumerate(y):
    ax.text(v + 3, i + .25, str(v), color='blue', fontweight='bold')
```

TAM = (Total Telehealth Market / Total Healthcare Market) * Budget for treating the specific disease

$$ TAM = \frac {Total\Telehealth\Market}{Total\Healthcare\Market}$$ $$ times Budget\for\treating\the\disease $$
